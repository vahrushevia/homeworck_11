```python
# f(x) = sin(x)^2 - cos(x)^2
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
import numpy as np
from sympy import Symbol, diff
from scipy.optimize import minimize_scalar
```

```python
# Определить корни
def func(x):
    return np.sin(x)**2 - np.cos(x)**2

root = fsolve(func, 0)
print(root)
```

```python
# Найти интервалы, на которых функция возрастает
def find_increasing_intervals(start, end, step):
    x = np.arange(start, end, step)
    y = np.sin(x)**2 - np.cos(x)**2

    increasing = np.diff(y) > 0
    intervals = []
    start = None
    for i, is_increasing in enumerate(increasing):
        if is_increasing:
            if start is None:
                start = i
        else:
            if start is not None:
                intervals.append((x[start], x[i]))
                start = None

    return intervals

print(find_increasing_intervals(-10, 10, 1))
```

```python
# Найти интервалы, на которых функция убывает
def find_decreasing_intervals(start, end, step):
    x = np.arange(start, end, step)
    y = np.sin(x)**2 - np.cos(x)**2

    decreasing = np.diff(y) < 0
    intervals = []
    start = None
    for i, is_decreasing in enumerate(decreasing):
        if is_decreasing:
            if start is None:
                start = i
        else:
            if start is not None:
                intervals.append((x[start], x[i]))
                start = None

    return intervals

print(find_decreasing_intervals(-10, 10, 1))
```

```python
# Построить график
x = np.linspace(-10, 10, 1000)
y = np.sin(x)**2 - np.cos(x)**2
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('f(x) = sin(x)^2 - cos(x)^2')
plt.grid()
plt.show()
```

```python
# Вычислить вершину
def f(x):
    return np.sin(x)**2 - np.cos(x)**2

result = minimize_scalar(lambda x: -f(x), bounds=(-10, 10), method='bounded')
print(result.x, f(result.x))
```

```python
# Определить промежутки, на котором f > 0

intervals1 = []
start = None
for i in np.arange(-10, 10, 1):
    if f(i) > 0 and start is None:
        start = i
    elif f(i) <= 0 and start is not None:
        intervals1.append((start, i))
        start = None
print(intervals1)
```

```python
# Определить промежутки, на котором f < 0

intervals2 = []
start = None
for i in np.arange(-10, 10, 0.01):
    if f(i) < 0 and start is None:
        start = i
    elif f(i) >= 0 and start is not None:
        intervals2.append((start, i))
        start = None

print(intervals2)
```
